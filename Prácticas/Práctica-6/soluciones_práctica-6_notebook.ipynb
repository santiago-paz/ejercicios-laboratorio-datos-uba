### **Ejercicio 16**

```python
# Cargamos nuevamente el dataset macro_full_columns.csv, usando la columna Anio como Ã­ndice
macroFull = pd.read_csv('macro_full_columns.csv', index_col='Anio')

# Ejercicio 16 (a)
# Creamos un nuevo dataset que contenga solo las columnas que terminan con "".oferta"

macro_oferta = macroFull.loc[:, macroFull.columns.str.endswith('.oferta')]
macro_oferta.info()
```

#### Como vemos ðŸ‘†, no hay datos faltantes en estas columnas

```python
# Ejercicio 16 (b)
# Creamos un dataframe X que contiene todas las columnas que terminan con ".oferta", excepto la columna "PBI_a_precios_de_mercado.oferta"

X = macroFull.loc[:, macroFull.columns.str.endswith('.oferta') & ~macroFull.columns.str.endswith('PBI_a_precios_de_mercado.oferta')]

# Creamos una Series con los valores de la columna PBI_a_precios_de_mercado.oferta

y = macroFull['PBI_a_precios_de_mercado.oferta']
```

```python
# Ejercicio 16 (c)
# Ajustamos un modelo de regresiÃ³n lineal ordinaria Ridge con X e y

from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(X, y)

# Generamos el vector de predicciones

y_pred = ridge.predict(X)
y_pred
```

```python
# Ejercicio 16 (d)

df_y_y_pred = pd.concat([y, pd.Series(y_pred).set_axis(y.index)], axis=1)
df_y_y_pred

# Graficamos los valores reales de PBI_a_precios_de_mercado.oferta contra los valores predichos

(
    so.Plot(x = df_y_y_pred.index, y = df_y_y_pred["PBI_a_precios_de_mercado.oferta"])
    .add(so.Dot(color="blue"))
    .add(so.Dot(color="red"), y = df_y_y_pred[0])
    .label(
        x="AÃ±o",
        y="PBI",
        title="PBI a precios de mercado",
    )

)
```
