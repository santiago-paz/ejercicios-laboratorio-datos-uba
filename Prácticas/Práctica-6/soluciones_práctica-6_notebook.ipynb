### **Ejercicio 1**

```python
import seaborn as sns
import seaborn.objects as so
import pandas as pd
import numpy as np

# Cargamos el dataset penguins

penguins = sns.load_dataset('penguins')

# Creamos un subconjunto de datos que contengan solo los pinguinos de la isla Biscoe, y que tengan un pico mayor de 48mm

penguins_biscoe = penguins[(penguins['island'] == 'Biscoe') & (penguins['bill_length_mm'] > 48)]
penguins_biscoe.head()
```

### **Ejercicio 2**

```python
# Creamos otro dataset de pinguinos macho Adelie que no sean de la isla de Biscoe

penguins_adelie = penguins[(penguins['species'] == 'Adelie') & (penguins['island'] != 'Biscoe')]
penguins_adelie.head()
```

### **Ejercicio 3**

```python
# Del dataset penguins nos quedamos con todas las variables excepto year, sex y body_mass_g

# Preguntar: No existe la variable 'year' en el dataset penguins
#penguins_subset = penguins.drop(columns=['year', 'sex', 'body_mass_g'])

penguins_subset = penguins.drop(columns=['sex', 'body_mass_g'])
penguins_subset.head()
```

### **Ejercicio 4**: Restablecer Ã­ndices

```python
# Ejercicio 4 (a)
# Eliminamos las filas con datos faltantes en el dataset penguins

penguins_clean = penguins.dropna()
penguins_clean.head()
```

#### Como vemos ðŸ‘†, los indices siguientes no son consecutivos, ya que hemos eliminado filas

```python
# Ejercicio 4 (b)
# Restablecemos el Ã­ndice del dataset penguins_clean

penguins_clean = penguins_clean.reset_index(drop=True)
penguins_clean.head()
```

```python
# Ejercicio 4 (c)
# Hacemos lo mismo encadenando los mÃ©todos dropna() y reset_index()

penguins_clean = penguins.dropna().reset_index(drop=True)
penguins_clean.head()
```

### **Ejercicio 5**: Renombrar columnas e Ã­ndices

```python
# Ejercicio 5 (a)
# Renombramos la columna "species" a "especies", en el dataset penguins

penguins_renombrada = penguins.rename(columns={'species': 'especies'})
penguins_renombrada.head()
```

```python
# Ejercicio 5 (b)
# Renombramos la variable flipper_length_mm a aleta_mm y la variable body_mass_g a peso_g en el dataset penguins_renombrada

penguins_renombrada = penguins_renombrada.rename(columns={'flipper_length_mm': 'aleta_mm', 'body_mass_g': 'peso_g'})
penguins_renombrada.head()
```

```python
# Ejercicio 5 (c)
# Renombramos el indice 0 a "cero" en el dataset penguins_renombrada

penguins_renombrada = penguins_renombrada.rename(index={0: 'cero'})
penguins_renombrada.head()
```

```python
# Ejercicio 5 (d)
# Pasamos todos los nombres de las columnas a mayÃºsculas en el dataset penguins_renombrada

penguins_renombrada.columns = penguins_renombrada.columns.str.upper()
penguins_renombrada.head()
```

```python
# Ejercicio 5 (e)

penguins = penguins.rename(index = np.sqrt)
penguins.head()
```
