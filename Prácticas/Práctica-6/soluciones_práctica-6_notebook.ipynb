## Operaciones con DataFrames y transformaciones de datos.

```python
import seaborn as sns
import seaborn.objects as so
import pandas as pd

# Ejercicio 1
# Cargamos el dataset penguins

penguins = sns.load_dataset('penguins')

# Creamos un subconjunto de datos que contengan solo los pinguinos de la isla Biscoe, y que tengan un pico mayor de 48mm

penguins_biscoe = penguins[(penguins['island'] == 'Biscoe') & (penguins['bill_length_mm'] > 48)]
penguins_biscoe.head()
```

```python
# Ejercicio 2
# Creamos otro dataset de pinguinos macho Adelie que no sean de la isla de Biscoe

penguins_adelie = penguins[(penguins['species'] == 'Adelie') & (penguins['island'] != 'Biscoe')]
penguins_adelie.head()
```

```python
# Ejercicio 3
# Del dataset penguins nos quedamos con todas las variables excepto year, sex y body_mass_g

# Preguntar: No existe la variable 'year' en el dataset penguins
#penguins_subset = penguins.drop(columns=['year', 'sex', 'body_mass_g'])

penguins_subset = penguins.drop(columns=['sex', 'body_mass_g'])
penguins_subset.head()
```

```python
# Ejercicio 4 (a)
# Eliminamos las filas con datos faltantes en el dataset penguins

penguins_clean = penguins.dropna()
penguins_clean.head()
```
