### **Carga de librerÃ­as y del dataset penguins**

```python
import seaborn as sns
import seaborn.objects as so
import pandas as pd
import numpy as np

penguins = sns.load_dataset('penguins')
```

### **Ejercicio 1**

```python
# Creamos un subconjunto de datos que contengan solo los pinguinos de la isla Biscoe, y que tengan un pico mayor de 48mm

penguins_biscoe = (penguins[
    (penguins['island'] == 'Biscoe') &
    (penguins['bill_length_mm'] > 48)])
penguins_biscoe.head()
```

### **Ejercicio 2**

```python
# Creamos otro dataset de pinguinos macho Adelie que no sean de la isla de Biscoe

penguins_adelie = penguins[(penguins['species'] == 'Adelie') & (penguins['island'] != 'Biscoe')]
penguins_adelie.head()
```

### **Ejercicio 4**

```python
# Ejercicio 4 (a)
# Eliminamos las filas con datos faltantes en el dataset penguins

penguins_clean = penguins.dropna()
penguins_clean.head()
```

```python
# Ejercicio 4 (b)
# Restablecemos el Ã­ndice del dataset penguins_clean

penguins_clean = penguins_clean.reset_index(drop=True)
penguins_clean.head()
```

```python
# Ejercicio 4 (c)
# Hacemos lo mismo encadenando los mÃ©todos dropna() y reset_index()

penguins_clean = penguins.dropna().reset_index(drop=True)
penguins_clean.head()
```

### **Ejercicio 5**

```python
# Ejercicio 5 (a)
# Renombramos la columna "species" a "especies", en el dataset penguins

penguins_renombrada = penguins.rename(columns={'species': 'especies'})
penguins_renombrada.head()
```

```python
# Ejercicio 5 (b)
# Renombramos la variable flipper_length_mm a aleta_mm y la variable body_mass_g a peso_g en el dataset penguins_renombrada

penguins_renombrada = penguins_renombrada.rename(
    columns={'flipper_length_mm': 'aleta_mm', 'body_mass_g': 'peso_g'})
penguins_renombrada.head()
```

```python
# Ejercicio 5 (c)
# Renombramos el indice 0 a "cero" en el dataset penguins_renombrada

penguins_renombrada = penguins_renombrada.rename(index={0: 'cero'})
penguins_renombrada.head()
```

```python
# Ejercicio 5 (d)
# Pasamos todos los nombres de las columnas a mayÃºsculas en el dataset penguins_renombrada

penguins_renombrada.columns = penguins_renombrada.columns.str.upper()
penguins_renombrada.head()
```

```python
# Ejercicio 5 (e)

penguins = penguins.rename(index = np.sqrt)
penguins.head()
```

#### Como vemos ðŸ‘†, aplica a cada Ã­ndice de la fila su raÃ­z cuadrada

```python
# Ejercicio 5 (f)
# Sumamos 1 a todos los Ã­ndices del dataset penguins

def suma_uno(x):
    return x + 1

penguins = penguins.rename(index = suma_uno)
penguins.head()
```

```python
# Ejercicio 5 (g)
# Utilizamos una funciÃ³n lambda para hacer lo mismo que en el ejercicio anterior

penguins = penguins.rename(index = lambda x: x + 1)
penguins.head()
```

```python
# Ejercicio 5 (h)
# Utilizamos una funciÃ³n lambda para pasar todos los nombres de las columnas a mayÃºsculas en el dataset penguins

penguins = penguins.rename(columns = lambda x: x.upper())
penguins.head()
```

### **Ejercicio 9**

```python
# Cargamos nuevamente el dataset penguins para resetear los cambios realizados en el ejercicio anterior

penguins = sns.load_dataset('penguins')

# Creamos un subconjunto de datos solo con las observaciones de pinguinos con aletas mayores a 200mm

penguins_mayores_200 = penguins[penguins['flipper_length_mm'] > 200]

# Nos quedamos solamente con las columnas que terminan con "_mm" en el dataset penguins

penguins_mayores_200 = penguins_mayores_200.loc[:, penguins_mayores_200.columns.str.endswith('_mm')]

# Reseteamos el Ã­ndice del dataset penguins_mayores_200

penguins_mayores_200 = penguins_mayores_200.reset_index(drop=True)
penguins_mayores_200.head()
```

### **Ejercicio 10**

```python
# Creamos un nuevo dataframe a partir de penguins que tenga solo las observaciones de la isla "Dream"

penguins_dream = penguins[penguins['island'] == 'Dream']

# Nos quedamos con las columnas "species" y todas las que comienzan con "bill" en el dataset penguins_dream

penguins_dream = penguins_dream.loc[:, ['species'] + list(penguins_dream.columns[penguins_dream.columns.str.startswith('bill')])]

penguins_dream.head()
```

### **Ejercicio 11**

```python
# Ejercicio 11 (a)
# Creamos una nueva columna "peso_en_kg" en el dataset penguins que contenga el peso en kilogramos

penguins['peso_en_kg'] = penguins['body_mass_g'] / 1000
penguins.head()
```

### **Ejercicio 12**

```python
# Cargamos nuevamente el dataset penguins para resetear los cambios realizados en el ejercicio anterior

penguins = sns.load_dataset('penguins')

# Agregamos una nueva columna con la mediana de la masa corporal de cada especie de pinguino en el dataset penguins, usando group_by

penguins['mediana_peso'] = penguins.groupby('species')['body_mass_g'].transform('median')

# Obtenemos filas al azar del dataset penguins

penguins.sample(10)
```
