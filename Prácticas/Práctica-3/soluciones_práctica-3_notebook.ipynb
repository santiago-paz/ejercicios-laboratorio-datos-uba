## Visualización de datos

### Ejercicio 1.a

```python
datosArgentina = gapminder[gapminder['country'] == 'Argentina']
(
so.Plot(data = datosArgentina , x = "year", y = "lifeExp")
.add(so.Line())
)
```

### Ejercicio 1.b

```python
paisesAmerica = gapminder[gapminder['continent'] == 'Americas']

(so.Plot(data = paisesAmerica,x = "year", y = "lifeExp", color="country")
.add(so.Line()))
```

### Ejercicio 1.c

```python
# Código 1
(
    so.Plot(data = paisesAmerica, x = "year", y = "lifeExp", group="country")
    .add(so.Lines(color="#bbca"))
    .add(so.Line(), so.PolyFit (1))
)
# Código 2, este es el código correcto para crear una curva de tendencia lineal  de la esperanza de vida en los países de América. La diferencia con el código 1 es que el primero genera una curva de tendencia lineal para cada país, mientras que el segundo genera una única curva de tendencia lineal para todos los países de América.
(
    so.Plot(data = paisesAmerica, x = "year", y = "lifeExp")
    .add(so.Lines(color="#bbca"), group="country")
    .add(so.Line(), so.PolyFit (1))
)
```
